```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial.csv', sep=';')
datos
```

```python
datos2= pd.read_csv('alquiler.csv', sep=';')
datos2.head(10)
```

```python
#agrupar locales con 1-2, 3-4, 5-6 y 7 o más recamaras
```

```python
clases = [0,2,4,6,100]
```

```python
# con base en la lista clases vamos a realizar los cortes para el filtro
# los cortes serian de esta manera
# del 0 hasta el 2, del 2 al 4 y asi sucesivamente
```

```python
cuartos = pd.cut(datos2.Cuartos,clases)
cuartos
```

```python
# los que aparecen en el grupo NaN es porque no hay un grupo definido para ellos
```

```python
pd.value_counts(cuartos)
```

```python
labels = ['1 y 2 cuartos', '3 y 4 cuartos', '5 y 6 cuartos', '7 cuartos o más']
```

```python
cuartos_lab = pd.cut(datos2.Cuartos,clases, labels = labels)
cuartos_lab
```

```python
# para incluir el 0 se agrego la propiedad 'include lowest = True'
cuartos = pd.cut(datos2.Cuartos,clases, labels = labels, include_lowest = True)
cuartos
```
